```python
%%writefile traversal.cpp

#include <iostream>
#include <queue>
#include <omp.h>

using namespace std;

int graph[10][10], visited[10];

// ---------- BFS ----------

void bfs(int start, int n)
{
    queue<int> q;

    visited[start] = 1;
    q.push(start);

    while(!q.empty())
    {
        int node = q.front();
        q.pop();

        cout << node << " ";

        #pragma omp parallel for
        for(int i = 0; i < n; i++)
        {
            if(graph[node][i] == 1 && !visited[i])
            {
                visited[i] = 1;

                #pragma omp critical
                q.push(i);
            }
        }
    }
}

// ---------- DFS ----------

void dfs(int node, int n)
{
    visited[node] = 1;

    cout << node << " ";

    for(int i = 0; i < n; i++)
    {
        if(graph[node][i] == 1 && !visited[i])
        {
            dfs(i, n);
        }
    }
}

// ---------- Main ----------

int main()
{
    int n, edges;

    cout << "Enter number of vertices: ";
    cin >> n;

    cout << "Enter number of edges: ";
    cin >> edges;

    cout << "Enter edges:\n";

    for(int i = 0; i < edges; i++)
    {
        int u, v;

        cin >> u >> v;

        graph[u][v] = 1;
        graph[v][u] = 1;
    }

    // ---------- BFS ----------

    for(int i = 0; i < n; i++)
        visited[i] = 0;

    double start = omp_get_wtime();

    cout << "\nBFS Traversal: ";

    bfs(0, n);

    double end = omp_get_wtime();

    cout << "\nBFS Execution Time = "
         << end - start << " seconds\n";

    // ---------- DFS ----------

    for(int i = 0; i < n; i++)
        visited[i] = 0;

    start = omp_get_wtime();

    cout << "\nDFS Traversal: ";

    dfs(0, n);

    end = omp_get_wtime();

    cout << "\nDFS Execution Time = "
         << end - start << " seconds\n";

    return 0;
}
```
